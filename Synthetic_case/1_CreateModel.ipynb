```python
## Create a test model

pos = 40
nLayers = 2

x =  np.linspace(0,pos,pos, endpoint=False)
sigm_1 = 20/1000
sigm_2 = 1200/1000

thk_1 = 2* np.ones(pos) # - 2*np.sin(np.pi*x/8)/8
thk_1[9:] = 2* np.ones(pos-9) - (5*x[:31]/100) 
thk_1[30:] = thk_1[30]

models = np.zeros((pos, nLayers*2-1))
models[:,0] = thk_1
models[:,1] = sigm_1
models[:,2] = sigm_2
```

```python
boundary = np.array([x, -thk_1]).T
```
