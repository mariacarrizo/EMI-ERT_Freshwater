## Upload data and sort

```python
OUT = np.load('data/data3D_slope10_40pos.pkl', allow_pickle=True)
```

```python
OUT
```

```python
OUT_sorted = OUT.sort_values(by='geom')
```

```python
OUT_sorted
```

```python
npos = int(len(OUT_sorted)/9)
```

```python
npos
```

```python
data_3D = np.zeros((npos-3,18))

pos=0
for p in range(-3,npos):
    #print(OUT_sorted.loc[(OUT_sorted.midpx > p+.06) & (OUT_sorted.midpx <p +1.06)])

    dat = np.hstack((OUT_sorted.loc[(OUT_sorted.midpx > p+.06) & (OUT_sorted.midpx <=p +1.06)]['op'],
                     OUT_sorted.loc[(OUT_sorted.midpx > p+.06) & (OUT_sorted.midpx <=p +1.06)]['ip'])) 
    if dat.size == 0:
        continue
    if dat.size == 18:
        data_3D[pos,:] = dat
        #print('pos:', pos)
        pos += 1
```

```python
np.save('data/data_3D_slope10_40pos_sorted', data_3D[1:41])
```
